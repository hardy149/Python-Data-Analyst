## 一、JSON格式

### 1. JSON文件

JSON全称为JavaScript Object Notation，是程序员常用的数据格式类型，文件是纯文本格式。

#### JSON的对象

JSON的对象与Python的字典相类似，都是由键值对构成：{"键1":"值1", "键2":"值2", ..., "键n":"值n"}

其不同之处在于：

Json的对象中只能由字符串作为键，且字符串必须被双引号包围（单引号不行）

Json的对象中的值，只能为以下几种数据类型：

字符串（“star”），数字（98），布尔值（ture），数组（[“hi”, 5]），对象（{“age”:25}），空值（null）

#### JSON的数组

JSON的对象与Python的列表相类似：[“xiaoming”, 71, {dady_xiaoming:17776573897}]

列表中的值可以为以下几种数据类型：

字符串（“star”），数字（98），布尔值（ture），数组（[“hi”, 5]），对象（{“age”:25}），空值（null）

## 补充：文件读写

### 文件的读写分三步走：打开文件，操作文件，关闭文件  

### 1. 打开与关闭文件

运用open函数来打开文件: open(file_name, mode, encoding=None)  

file_name是文件路径（绝对路径&相对路径）  

mode是打开文件的方式，最常用的有r（只读），w(只写)，a(追加)  

encoding是编码方式，默认为None，常见编码方式为utf-8

注：open( )方法的返回值是一个file对象，可以将它赋值给一个变量。

运用close函数来关闭文件: file.close( )

注：这里的file是一个对象，使用对象调出函数，来操作关闭。

#### 打开与关闭文件

```python
# 打开一个文件（将打开的文件赋值给变量f）
f = open("/Users/hardy/Desktop/Python_file/Python_DataAnalyst/Data_FileFomat/poem.txt", "r")
f.close()
```

#### 补充：如何找到文件路径

把文件拖入终端中，即可得到文件的绝对路径

### 2. 操作文件

对文件的操作主要是读写文件中的内容

运用read函数读取文件：read（ ）

运用write函数写文件：write（“写入的内容”）

注：要读文件需要在“r”(只读)模式下打开文件，要写文件需要在“w”(只写)模式下打开文件

#### （1）写入文件

```python
f = open("/Users/hardy/Desktop/Python_file/Python_DataAnalyst/Data_FileFomat/poem.txt", "w")
f.write('''
《酬乐天扬州初逢席上见赠》
刘禹锡
巴山楚水凄凉地，二十三年弃置身。
怀旧空吟闻笛赋，到乡翻似烂柯人。
沉舟侧畔千帆过，病树前头万木春。
今日听君歌一曲，暂凭杯酒长精神。
''')
f.close()
```

#### （2）读取文件

```python
f = open("/Users/hardy/Desktop/Python_file/Python_DataAnalyst/Data_FileFomat/poem.txt", "r")
print(f.read())
f.close()
```

### 3. 运用with打开和操作文件

运用with关键字打开文件，可以避免open打开文件在操作过程可能出现的异常或错误，或者最后忘执行close，导致非正常关闭进而使得文件泄露、破坏的问题。

运用with打开文件：with open (file_name, mode)  as f

```python
with open("/Users/hardy/Desktop/Python_file/Python_DataAnalyst/Data_FileFomat/poem.txt", "r") as f:
    print(f.read())
```

#### 运用with打开多个文件

```python
with open("/Users/hardy/Desktop/Python_file/Python_DataAnalyst/Data_FileFomat/poem.txt", "r") as f1, open("/Users/hardy/Desktop/Python_file/Python_DataAnalyst/Data_FileFomat/foo.txt", "w") as f2:
    f2.write("python是我学习的第一门编程语言。\n我会学好python的！")
    print(f1.read())
```
