# 一、数据评估标准

### 1. 干净数据的标准
（1）每列是一个变量；（2）每行是一个观察值；（3）每个单元格是一个值

### 2. 混乱数据的表现

（1）丢失数据；（2）重复数据；（3）不一致数据（不同数据值实际指代同一目标）；（4）无效或错误数据

### 3.评估丢失数据

```python
#找出各列非空缺值————isnull
df_amazon.isnull()
```

```python
#对各列空缺值进行求和
df_amazon.isnull().sum()
```

```python
#对特定列空缺值进行求和
df_amazon["rating_count"].isnull().sum()
```

```python
#提取存在空缺值的行
df_amazon[df_amazon["rating_count"].isnull()]
```

注：也可以运用info方法列出非空缺值数量与行数进行对比

### 4.评估重复数据

```python
#查看dataframe（特定列）中的重复数据————duplicated
df_amazon["discount_percentage"].duplicated()
```

注：只有后面出现的行与前面某行的值完全一致时，才为Ture

```python
#对dataframe（特定列）中的重复数据求和
df_amazon["discount_percentage"].duplicated().sum()
```

```python
#查看dataframe（特定多列）的重复数据——即两列元素完全一致才会被认定为重复
df_amazon.duplicated(["discount_percentage", "rating"])
```

```python
#查看dataframe（特定多列）的重复数据个数求和
df_amazon.duplicated(["discount_percentage", "rating"]).sum()
```

```python
#找出特定多行重复的数据
df_amazon[df_amazon.duplicated(["discount_percentage", "rating"])]
```

### 5.评估不一致数据

```python
#查看某列下面所有值的种类及次数————value_counts
df_amazon["discount_percentage"].value_counts()
```

注：value_counts会把某列出现的元素全部展示出来，并对其进行计数

### 6.评估无效错误数据

```python
#对某行排序查找最大最小数据查看错误数据————sort
df_amazon["rating"].sort_values()
```

```python
#对某行排序进行描述统计查看错误数据————describe
df_amazon["rating"].describe()
```
