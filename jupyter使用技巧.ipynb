# 关于jupyter notebook的学习笔记

## 1. 编辑&命令模式

jupyter notebook中包含两种模式：编辑模式&命令模式；

在编辑模式下，主要是往cell（单元格）中输入代码或释义；

在命令模式下，主要是对cell进行增加、删除、调整位置、运行等操作。

## 2. cell的类型

（1）code模式：编辑代码，运行后显示代码运行结果；

（2）markdown模式：编写markdown文档，运行后输出markdown格式文档；

（3）raw模式：普通文本，运行不会输出结果。

## 3. 命令栏

（1）File: 创建新的 Notebook、打开新的界面、拷贝当前 Notebook、重命名 Notebook、保存还原点、恢复到指定还原点、查看 Notebook 预览、下载 Notebook 、关闭 Notebook。

(2) Edit: 主要是对 Cells 剪切、复制、删除、向上/向下移动等。

(3) Insert: 插入 Cell，插入方式有两种：一种是在当前 cell 上方插入，一种是在当前 cell 下方插入。

(4) Kernel: 主要包含了中断 kernel、重启 kernel、重启 kernel 并清除输出、重启 kernel 并运行所有 cell、重连 kernel、关闭 kernel、改变 kernel 类型等。

(5) Help: 主要包含了用户交互引导、键盘快捷键、修改键盘快捷键、Notebook 帮助、Markdown 帮助、Jupyter-notebook-extensions 帮助、Python 帮助、IPython 帮助、Numpy 帮助、Scipy 帮助、Matplotlib 帮助、Sympy 帮助、pandas 帮助等。

## 4. 快捷键

A 键 在选中单元格上方插入单元格

B 键 在选中单元格下方插入单元格

X 键 剪切选中单元格

C 键 复制选中单元格

V 键 粘贴选中单元格

Z 键 撤销已删除的单元格

D 键 点按两次删除单元格

F 键 弹出 查找和替换

Y 键 将单元格转换为code模式

M 键 将单元格转换为markdown模式

R 键 将单元格转换为raw模式

shift + up 键 向上拓展选中单元格

shift + down 键 向下拓展选中单元格

command + enter 键 运行选中单元格

shift + enter 键 运行当前单元格，并跳到下一个单元格

1/2/3/4/5/6 键 将单元格转变为一/二/三/四/五/六级标题

‘* 斜体字 *’ = *斜体字*

‘~~ 删除 ~~’ =  ~~删除~~
